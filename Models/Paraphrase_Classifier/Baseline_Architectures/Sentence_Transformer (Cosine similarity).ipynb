### Baseline: Cosine similarity of Sentence Embeddings

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('sentence-transformers/bert-base-nli-mean-tokens')
```

```python
model2 = SentenceTransformer("sentence-transformers/paraphrase-albert-small-v2")
```

```python
model3 = SentenceTransformer("sentence-transformers/paraphrase-multilingual-mpnet-base-v2")
```

```python
# model4 = SentenceTransformer("sentence-transformers/all-MiniLM-L12-v2")
```

```python
sentences = ["I like you", "I hate you"]
embeddings = model.encode(sentences)
print(embeddings)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
emb0 = embeddings[0].reshape(-1,1)
emb0 = np.swapaxes(emb0, 0, 1)
# print(emb0.shape)
emb1 = embeddings[1].reshape(-1,1)
emb1 = np.swapaxes(emb1, 0, 1)
```

```python
ans = cosine_similarity(emb0,emb1)
print(ans)
```

```python
from tqdm import tqdm
```

```python
onelabels=[]
zerolabels=[]
all_labels = []
for i in tqdm(range(len(dataset['test']))):
  x = dataset['test'][i]
  # print(x)
  sentences = [x['sentence1'], x['sentence2']]
  gt_label = x['label']
  embeddings = model.encode(sentences)
  emb0 = embeddings[0].reshape(-1,1)
  emb0 = np.swapaxes(emb0, 0, 1)
  # print(emb0.shape)
  emb1 = embeddings[1].reshape(-1,1)
  emb1 = np.swapaxes(emb1, 0, 1)
  pred_label = None
  cos_sim = cosine_similarity(emb0,emb1)[0]
  if gt_label==1:
    onelabels.append(cos_sim[0])
  else:
    zerolabels.append(cos_sim[0])
  all_labels.append(cos_sim[0])
```

```python
# len(onelabels)
```

```python
# len(zerolabels)
```

```python
from statistics import mean
onemean = mean(onelabels)
zeromean = mean(zerolabels)
# btwmean = (onemean+zeromean)/2
print(onemean)
print(zeromean)
# print(btwmean)
```

```python
pred_labels = [int(x/onemean) for x in all_labels]
# print(pred_labels[:10])
# print(all_labels[:10])
```

```python
sum_array = [sum(x) for x in zip(dataset['test'][:]['label'],pred_labels)]
same_value_array = [i for i in sum_array if i != 1]
correct = len(same_value_array)
```

```python
print("Test Set Accuracy :", (correct/len(dataset['test'])*100), "%")
```

```python
correct = 0
for i in tqdm(range(len(dataset['validation']))):
  x = dataset['validation'][i]
  # print(x)
  sentences = [x['sentence1'], x['sentence2']]
  gt_label = x['label']
  embeddings = model.encode(sentences)
  emb0 = embeddings[0].reshape(-1,1)
  emb0 = np.swapaxes(emb0, 0, 1)
  # print(emb0.shape)
  emb1 = embeddings[1].reshape(-1,1)
  emb1 = np.swapaxes(emb1, 0, 1)
  pred_label = None
  if cosine_similarity(emb0,emb1)[0]>=onemean:
    pred_label = 1
  else:
    pred_label = 0
  if gt_label == pred_label:
    correct = correct + 1

print(correct/len(dataset['validation']))
```

```python
print("Validation Set Accuracy :", (correct/len(dataset['validation'])*100), "%")
```
